```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
columns = ['EIN', 'NAME']
application_df = application_df.drop(columns=columns)
application_df
```

```python
application_df['INCOME_AMT'].value_counts()
```

```python
# Define the mapping of income ranges to bins
income_bins = {
    '10M-50M': '10M+',
    '5M-10M': '10M+',
    '50M+': '10M+'
}

# Apply the binning to the 'INCOME_AMT' column
application_df['INCOME_AMT'].replace(income_bins, inplace=True)
application_df
```

```python
application_df['APPLICATION_TYPE'].value_counts()
```

```python
application_df['ORGANIZATION'].value_counts()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
 application_df['SPECIAL_CONSIDERATIONS'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
value_counts = application_df['APPLICATION_TYPE'].value_counts()
value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff = 528
application_types_to_replace = []
for app_type, count in value_counts.items():
    if count < cutoff:
        application_types_to_replace.append(app_type)
        
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
filtered_counts = classification_counts[classification_counts > 1]
filtered_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classification_types_to_replace`
cutoff_value = 1883
classification_types_to_replace = []
for app_type, count in classification_counts.items():
    if count < cutoff_value:
        classification_types_to_replace.append(app_type)

# Replace in dataframe
for cls in classification_types_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
d_columns = ['IS_SUCCESSFUL','STATUS','SPECIAL_CONSIDERATIONS']
y = application_df['IS_SUCCESSFUL']
x = application_df.drop(columns= d_columns )
x = pd.get_dummies(x)
x[:5]
```

```python
# Split our preprocessed data into our features and target arrays
# Split the preprocessed data into a training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
x_scaler = scaler.fit(x_train)

# Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
```

## Compile, Train and Evaluate the Model

```python
#Find Input Dimensions
dimensions = len(x.columns)
dimensions
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn_model = tf.keras.models.Sequential()

# First hidden layer, Second hidden layer, Output layer
nn_model.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim = dimensions))
nn_model.add(tf.keras.layers.Dense(units=60, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=60, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=30, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=20, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(x_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(x_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save('AlphabetSoupCharity_Optimization.h5')
```
